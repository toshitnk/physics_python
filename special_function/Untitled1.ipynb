# $\Gamma$関数，$B$関数

- $\Gamma$関数

```python
%display typeset

plot(gamma(x),(x,-5,4),ymin=-5, ymax=5)
```

片対数で．

```python
plot_semilogy(gamma(x), (x,-5,10))
```

### $1/\Gamma(x)$
マイナスの方で振動がでっかくなっていく．

```python
plot(1/gamma(x), (x,-5,5)).show()
plot_semilogy(1/gamma(x), (x, -10, 10))
```
