## 2準位系の確率

```python
beta = np.linspace(0, 5, 1000)
p1 = 1 / (1 + np.exp(-beta))
p2 = np.exp(-beta) / (1 + np.exp(-beta)) 

plt.xticks([0, 5], ["high temp.", "low temp."])
plt.yticks([0,0.5,1])
plt.xlabel(r"$\beta$", fontsize=18)
plt.ylabel("probability", fontsize=18)
plt.plot(beta, p1, label="low energy")
plt.plot(beta, p2, label="high energy")
plt.legend()
```

## sinh, cosh, tanh

```python
x = np.linspace(-np.pi, np.pi, 1000)
sinh = np.sinh(x)
cosh = np.cosh(x)
tanh = np.tanh(x)

# plt.ylim(-3,3)

plt.plot(x, sinh, label=r"$\sinh(x)$")
plt.plot(x, cosh, label=r"$\cosh(x)$")
plt.plot(x, tanh, label=r"$\tanh(x)$")

plt.legend()
```

# 結晶の比熱
- Einstein model
- Debye model
- 最も簡単な3次元モデル

```python
x = np.linspace(0,10,1000)

CE = 1 / (x**2) * ( 1 / (np.sinh(1/(2 * x) ))**2)

plt.plot(x, CE)
```

```python
def deb(x):
    if x == 0:
        return 0
    return x ** 4 * (1 / (2 * np.sinh(x/2))**2)

DE = []
for x in np.linspace(0,10,1000):
    d, err =  integrate.quad(deb, 0., 1./x)
    DE.append(d*(x**3))

plt.plot(np.linspace(0,10,1000), DE)
```
