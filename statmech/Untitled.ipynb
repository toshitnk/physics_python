```python
import numpy as np
import matplotlib.pyplot as plt

m = 0
e = np.linspace(-10**4+m , 10**4+m, 100)

T_list = np.array([10, 100, 300, 500, 1000])

for T in T_list:
    fd = 1 / (np.exp((e-m)/T)+1)
    plt.plot(e, fd, label=r"$T=$%i"%T)
plt.legend()
```
