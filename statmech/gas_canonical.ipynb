## Maxwell-Boltzmann分布
- 分子の速度の分布関数．
- 温度依存性をみる．

```python
x = np.linspace(0,50,1000)
Thermo = [50, 100, 300, 1000]

for i in range(0,len(Thermo)):
    T = Thermo[i]
    mb = (1/T) ** (3/2) * x ** 2 * np.exp(-x**2/T)
    plt.plot(x, mb, label=r"$T=%i$"%(Thermo[i]))
    plt.legend()
```

### 古典Boltzmann factorの温度依存性

```python
x = np.linspace(0,30,1000)

V = [1,5,10]

for i in range(0,len(V)):
    bol = np.exp(-V[i]/x)

    plt.plot(x, bol, label=r"$V=%i$"%(V[i]))    
    plt.legend()

```

# 高分子の弾性
### Langevin関数（長さの期待値）

```python
x = np.linspace(-10, 10, 1000)

L = 1 / np.tanh(x) - 1 / x

plt.plot(x, L)
```
